Dogecoin Price Prediction with Machine Learning

Dogecoin is a cryptocurrency that have the same features like Ethereum or Bitcoin.

In this article, we will be implementing a machine learning model which can predict the pattern or forecast the price of the coin in the upcoming days.
Let us now move towards the implementation of price prediction.

---



```python
# Import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
```

Load the dataset

```python
data = pd.read_csv("DOGE-USD.csv")
data.head()
```

Now, let's check for correlation of the dataset.

```python
data["Date"] = pd.to_datetime(data["Date"], infer_datetime_format=True)
data.set_index("Date", inplace=True)
data.isnull().any()
```

```python
data.corr()
```

Let's check the presence of missing values.

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

Now, let's check for the statistical analysis of the data using the describe() method.

```python
data.describe()
```

Now, we want to analyze the closing price as we need it to perform the prediction

```python
plt.figure(figsize=(16, 8))
x = data.groupby("Date")["Close"].mean()
x.plot(linewidth=2.5, color="b")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Date vs Close of 2021")
```

The column "Close" is our predicted feature. We are taking different factors for our own calculation and naming them suitably. Also, we are checking each factor while correlating with the 'Close' column while sorting it in descending order.

```python


data["gap"] = (data["High"] - data["Low"]) * data["Volume"]
data["y"] = data["High"] / data["Volume"]
data["z"] = data["Low"] / data["Volume"]
data["a"] = data["High"] / data["Low"]
data["b"] = (data["High"] / data["Low"]) * data["Volume"]
abs(data.corr()["Close"].sort_values(ascending=False))
```

By, observing the correlating factors, we can choose a few of them. We are excluding High, Low, and Open as they are highly correlated from the beginning.

```python


data = data[["Close", "Volume", "gap", "a", "b"]]
data.head()
```

Introducing the ARIMA model for Time Series Analysis. ARIMA stands for autoregressive integrated moving average model and is specified by three order parameters: (p, d, q) where AR stands for Autoregression i.e. p, I stands for Integration i.e. d, MA stands for Moving Average i.e. q. Whereas, SARIMAX is Seasonal ARIMA with exogenous variables.

```python


df2 = data.tail(30)
train = df2[:11]
test = df2[-19:]
```

```python
print(train.shape, test.shape)
```

Model Development

```python

from statsmodels.tsa.statespace.sarimax import SARIMAX
model = SARIMAX(endog=train["Close"], exog=train.drop(
    "Close", axis=1), order=(2, 1, 1))
results = model.fit()
print(results.summary())
```

Now, observe the prediction.

Now, observe the prediction in time series.

```python

start = 11
end = 29
predictions = results.predict(
    start=start,
    end=end,
    exog=test.drop("Close", axis=1))
predictions
```

```python

test["Close"].plot(legend=True, figsize=(12, 6))
predictions.plot(label='TimeSeries', legend=True)
```
